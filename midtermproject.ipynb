## Exploring Relationships

```python
correlated_wine_qualities = RW_wine_df.corr(numeric_only=True)

plt.figure(figsize = (8, 8))

sns.heatmap(correlated_wine_qualities, annot=True, cmap="coolwarm")
plt.title("Wine Correlations")
plt.show()
```

```python
red_wine_qualities = red_wine_df.corr(numeric_only=True)

plt.figure(figsize = (8, 8))

sns.heatmap(red_wine_qualities, annot=True, cmap="coolwarm")
plt.title("Red Wine Correlations")
plt.show()
```

Note that above, on the Red Wines only chart, Density and Fixed Acidity are MUCH more strongly correlated (0.67) than on the White Wines only chart (0.27).
On both charts as well, the correlation between Density and Residual Sugar vary greatly, Red Wines at 0.36 and White Wines at 0.84

```python
white_wine_qualities = white_wine_df.corr(numeric_only=True)

plt.figure(figsize = (8, 8))

sns.heatmap(white_wine_qualities, annot=True, cmap="coolwarm")
plt.title("White Wine Correlations")
plt.show()
```

In Graph above you can see the relationship between Citric Acid and Volatile Acidity gradually shifting with each quality rating increasing. Volatile Acidity starts above Citric Acid, and by the 9 rating it ends up having quite a bit of a greater value than the Volatile Acidity. Additionally, chlorides are significantly at their lowest at the highest quality rating.

Chlorides and why lower levels tend to lead to better quality:
1. Moderate salinity can add complexity, but higher levels can make a wine taste "salty" rather than "mineral", detracting from the fruit forward profile that you'd expect from high quality wine -- the excess salinity can essentially mask delicate aromas and flavors.
2. Low-chloride wines are generally produced from healthier soils and better irrigated fresh water

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Feature Correlation with Wine Quality", fontsize=16, fontweight="bold")

for ax, col, color, ylim in zip(axes, ["chlorides", "citric acid"], ["blue", "red"], [(0, 0.3), (0, 1)]):
    sns.regplot(
        data=RW_wine_df,
        x="quality",
        y=col,
        ax=ax,
        color=color,
        scatter_kws={"alpha": 0.1, "s": 40},
        line_kws={"linewidth": 2}
    )
    corr = RW_wine_df[col].corr(RW_wine_df["quality"])
    ax.set_title(f"{col.title()} vs Quality", fontsize=13)
    ax.set_xlabel("Quality", fontsize=11)
    ax.set_ylabel(col.title(), fontsize=11)
    ax.set_ylim(ylim)

plt.tight_layout()
plt.show()
```

Clear distinction of greater alcohol content in both reds and whites leading to better quality scores, let's take a look at it on a lineplot
